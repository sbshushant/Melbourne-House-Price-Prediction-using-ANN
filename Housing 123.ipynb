```python
data1=data.dropna()
```

```python
x = data1.iloc[:,0:13].values
y=data1.iloc[:,13].values
```

```python
# splitting the training & test sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
mc=MinMaxScaler()
x_train1 = mc.fit_transform( x_train)
X_test1 = mc.fit_transform( x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
# Initialize the model
model=Sequential()

# Build input layer & Hidden layer
model.add(Dense(units=40, activation='relu', input_dim=13))
model.add(Dropout(rate=0.1))
```

```python
# Building second layer
model.add(Dense(units=20,activation ='relu'))
model.add(Dropout(rate=0.2))
```

```python
#Adding outer layer
model.add(Dense(units = 1, activation='linear'))
```

```python
# compile the model
model.compile(optimizer='adam', loss='mae')
```

```python
# fit model 
model.fit(x_train,y_train,epochs=100)
```
